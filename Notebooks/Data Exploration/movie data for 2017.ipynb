# Data Wrangling

Since we have collected data through 2016 in data.csv, I will only be extracting 2017 data from meta dataset. 

```python
meta_2017 = meta.loc[meta['year'] == 2017, 
                    ['genres', 'id', 'title', 'year']]
meta_2017.head()
```

```python
# merging the credits and meta_2017 dataframes
meta_2017['id'] = meta_2017['id'].astype(int)

data = pd.merge(meta_2017, credits, on='id')
data.head()
```

```python
data.genres.dtype
```

Using the ast library to evaluate the string as a python expression.

```python
data['genres'] = data['genres'].apply(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].apply(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].apply(lambda x: ast.literal_eval(x))
```

Extracting the information from genres, cast, crew

```python
def make_genresList(x):
    
    gen = []
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return ' '.join(gen)
```

```python
data['genres_list'] = data['genres'].apply(lambda x: make_genresList(x))
data['genres_list']
```

```python
def get_actor1(x):
    
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].apply(lambda x: get_actor1(x))
data['actor_1_name']
```

```python
def get_actor2(x):
    
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].apply(lambda x: get_actor2(x))
data['actor_2_name']
```

```python
def get_actor3(x):
    
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].apply(lambda x: get_actor3(x))
data['actor_3_name']
```

```python
def get_directors(x):
    
    directors = []
    for i in x:
        if i.get('job') == 'Director':
            directors.append(i.get('name'))
    if directors == []:
        return np.NaN
    else:
        return " ".join(directors)
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
data['director_name']
```

```python
movie = data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie
```

```python
movie.isna().sum().sort_values(ascending=False)
```

```python
# dropping all the rows with any NaN values
movie = movie.dropna(axis=0, how='any')
```

Making changes to match our other dataset with movies thru 2016 (data.csv)

```python
movie = movie.rename(columns={'genres_list':'genres', 'title':'movie_title'})
movie
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

We will now append this 2017 movie df to our 2016 data.csv 

```python
data_thru_2016 = pd.read_csv('data.csv')
```

```python
data_thru_2017 = data_thru_2016.append(movie)
```

Making a column with a combination of all the information which will be useful later for our recommender system. 

```python
data_thru_2017['comb'] = data_thru_2017['actor_1_name'] + ' ' + data_thru_2017['actor_2_name'] + ' ' + \
    data_thru_2017['actor_3_name'] + ' ' + \
    data_thru_2017['director_name'] + ' ' + data_thru_2017['genres']
data_thru_2017
```

```python
data_thru_2017.drop_duplicates(subset ="movie_title", keep = 'last', inplace = True)
```

```python
data_thru_2017.to_csv('data_2017.csv', index=False)
```
